## Data Set
This is an artificial dataset that contains patient id, age, sex, city, language, ever_smoker, current_smoker, and etc.

```python
patient = pd.read_csv('../artificial_patient_data.csv', header=[0])
cols = patient.columns.tolist()

patient.head()
```

## Visualizations for Patient Demographics

### Age and Race Distribution

```python
alt.hconcat(
    alt.Chart(patient).mark_bar().encode(
        alt.X("age:Q", bin=True),
        y='count()',
    ).properties(
        height=300
    ),
    alt.Chart(patient).mark_bar().encode(
        alt.X("sex:N"),
        y='count()',
    ),
    alt.Chart(patient).mark_bar().encode(
        alt.Y("race:N", sort='-x'),
        x='count()',
    ).properties(
        height=300
    )
)
```

### Distribution of States

```python
alt.Chart(patient).mark_bar().encode(
    alt.X("state:N", sort="-y"),
    y='count()',
)
```
